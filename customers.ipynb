## **FunDrive Car Insurance Customer Lifetime Value**

## **Business Problem Understanding**

### **Context**

Customer Lifetime Value (CLV), adalah salah satu indikator yang sangat penting bagi perusahaan dalam memahami seberapa berharga kontribusi pelanggan terhadap bisnis mereka. `CLV adalah sebuah metrik yang digunakan untuk mengukur berapa banyak keuntungan yang dapat dihasilkan dari hubungan jangka panjang dengan seorang pelanggan, sekaligus mempertimbangkan biaya terkait dengan mempertahankan pelanggan tersebut.`

Pemahaman CLV membantu perusahaan dalam menentukan efisiensi upaya pemasaran mereka dalam mempertahankan pelanggan yang berharga. Dengan mengetahui berapa banyak keuntungan yang dapat diperoleh dari setiap pelanggan dalam jangka panjang, perusahaan dapat `menilai apakah biaya pemasaran untuk mempertahaankan pelanggan yang dikeluarkan sebanding dengan nilai CLV yang dihasilkan.`

Selain itu, CLV membantu perusahaan dalam merencanakan strategi bisnis jangka panjang. Dengan memahami bagaimana pelanggan akan berubah seiring berjalannya waktu, perusahaan dapat mengalokasikan sumber daya mereka secara lebih efisien, mengidentifikasi peluang untuk meningkatkan loyalitas pelanggan, dan mengantisipasi perubahan dalam preferensi pelanggan. 

`Contohnya`, pada konteks data asuransi mobil ini, dalam jangka waktu kontrak asuransi, preferensi pelanggan/client mungkin berubah dan terdapat kemungkinan client untuk beralih ke asuransi lain setelah kontrak asuransi habis. Dengan mengetahui CLV, perusahaan dapat mengakomodir masalah tersebut melalui strategi pemasaran jangka panjang dan juga perusahaan dapat memfokuskan strategi tersebut secara lebih tepat sasaran kepada client yang bener-benar berharga atau yang memiliki CLV tinggi.

Dan juga, CLV dapat membantu dalam pengambilan keputusan tentang investasi dalam pengembangan produk atau layanan baru. `Contohnya` seperti renew offer asuransi. Perusahaan dapat memprioritaskan investasi mereka berdasarkan potensi CLV jangka panjang dari pelanggan yang diharapkan akan menikmati produk atau layanan tersebut.

`jadi pada intinya, dalam konteks bisnis asuransi mobil di data ini, CLV dapat membantu perusahaan untuk mengkalkulasi dan menjaga agar biaya mempertahankan pelanggan (retain) tidak melebihi nilai CLV. Karena jika biaya retain melebihi nilai CLV, perusahaan akan mengalami kerugian. Dan juga CLV membantu perusahaan untuk menentukan skala prioritas untuk client mana yang perlu di retain dengan memberikan offer, service atau treatment tertentu.`


**Problem Statement**

Dari konteks yang telah dijelaskan diatas, dapat diketahui bahwa CLV adalah tolak ukur yang penting dalam menghindari kerugian akibat biaya retain yang tidak tepat sasaran. `Namun, salah satu tantangan terbesar bagi perusahaan asuransi kendaraan seperti FunDrive Insurance ini adalah memprediksi nilai CLV itu sendiri.` Setiap perusahaan memiliki cara dan faktor yang berbeda dalam menentukan CLV nya, hal tersebut bergantung dari bagaimana proses bisnisnya. Misalkan, untuk menentukan CLV customer supermarket akan berbeda dengan menentukan CLV client asuransi. `Nilai fitur yang berbeda dari tiap clien juga cenderung akan menghasilkan CLV yang berbeda, sehingga hal ini menjadi permasalahan bagi FunDrive Insurance untuk memprediksi nilai CLV dengan model terbaik dan akurat.`

**Goals**

Berdasarkan permasalahan tersebut, FunDrive Insurance perlu memiliki `'tool machine learning'` yang dapat memprediksi serta membantu perusahaan untuk dapat `memprediksi nilai Customer Lifetime Value.` Sehingga perusahaan dapat `menghindari kerugian` dari biaya retain yang tidak sesuai dan juga perusahaan tahu mana client yang perlu di retain baik melalui pemberian produk, fasilitas, dan pemasaran tertentu yang lebih tepat sasaran.
Oleh karena itu, `melalui model machine learning yang di bangun dalam proses ini diharapkan dapat menghasilkan model prediksi CLV yang memiliki tingkat error yang kecil.`

**Stake Holder**

Hasil prediksi akan digunakan oleh Tim Marketing dalam menentukan biaya promosi dan retain client.

**Metric Evaluation**

Evaluasi metrik yang akan digunakan adalah `RMSE, MAE, dan MAPE`, di mana RMSE adalah akar dari rata-rata error yang dikuadratkan, MAE adalah rataan nilai absolut dari error, sedangkan MAPE adalah rataan persentase error yang dihasilkan oleh model regresi. Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi Customer Lifetime Value sesuai dengan limitasi fitur yang digunakan.

## **DATA PREPROCESSING**

```python
ordinal_mapping = [
    {'col':'Coverage', 'mapping':{'Basic':1, 'Extended':2, 'Premium':3}},
    {'col':'Education', 'mapping':{'High School or Below':1, 'College':2, 'Bachelor':3, 'Master':4, 'Doctor':5}} 
] 
```

```python
# transformer 

pipe_ordinal_scale = Pipeline([
    ('ordinal', OrdinalEncoder(cols=['Coverage', 'Education'], mapping=ordinal_mapping)),
    ('scaler', RobustScaler())
])

transformer = ColumnTransformer([
    ('robust', RobustScaler(), ['Number of Policies', 'Monthly Premium Auto', 'Total Claim Amount', 'Income']),
    ('pipe_ordinal_scale', pipe_ordinal_scale, ['Coverage', 'Education']),
    ('onehot', OneHotEncoder(drop='first'), ['Vehicle Class','Renew Offer Type', 'EmploymentStatus', 'Marital Status']),
], remainder='passthrough')

transformer
```

Data preprocessing yang dilakukan yaitu:
1. Melakukan scale pada fitur numerikal dengan menggunakan robust scaler. Robust scaler digunakan untuk menormalkan atau mengubah skala fitur numerik dalam sebuah dataset yang memiliki banyak outlier seperti pada data ini.
2.  lalu melakukan encoding pada fitur kategorikal. Encoding dilakukan untuk mengubah nilai kategorikal menjadi angka, karena machine learning hanya bisa membaca angka. Encoding yang dilakukan ada dua yaitu, ordinal encoder untuk fitur 'Coverage dan Education' karena value dalam kedua variable ini memiliki urutan atau tingkatan. Sedangkan variable kategorikal lainnya di encoding dengan metode onehot encoder karena bersifat nominal atau tidak memiliki urutan, dan unik nya tidak terlalu banyak.

## **MODELING**

- Setelah melalui tahap preprocessing, maka langkah selanjutnya adalah melalukan modeling dengan menentukan terlebih dahulu variable X dan Y nya.
- variable Y atau target dalam data ini adalah Customer Lifetime Value. dan variable sisanya adalah variable X atau fitur yang digunakan untuk memprediksi target (CLV). 

### Define X Y

```python
X = df.drop(columns=['Customer Lifetime Value'])
y = df['Customer Lifetime Value']
```

### Train Test Split

Setelah menentukan fitur dan target, hal yang selanjutnya dilakukan adalah membagi data menjadi dua bagian yaitu dataset train dan data set test. data train digunakan untuk melatih model dan melakukan proses belajar sedangkan data set test digunakan untuk menguji model terbaik yang telah dilatih di data set train. Data set test diambil secara rabdom sebanyak 25%.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    test_size=0.25,
    random_state=42
)
```

### Model Benchmarking

Setelah melalikan train test split, maka selanjutnya kita dapat melakukan model benchmarking. Karena target dalam data ini adalah data numerikal maka model/algoritma yang digunakan adalah regresi. Terdapat beberapa algoritma regresi yang coba di benchmark yaitu, Linear regression, KNN Reg, Decision Tree Reg. Lalu algoritma regularisasi dengan lasso, ridge dan elastict net. Serta dengan 4 model ensamble yaitu XGboost, Random Forest, Gboost dan Catboost.

```python
linear_reg = LinearRegression()
knn_reg = KNeighborsRegressor()
tree_reg = DecisionTreeRegressor(random_state=42)
lasso = Lasso()
ridge = Ridge()
enet = ElasticNet()

# 4 ensemble model
xgb_reg = XGBRegressor(random_state=42)
rf_reg = RandomForestRegressor(random_state=42)
gboost_reg = GradientBoostingRegressor(random_state=42)
catboost= CatBoostRegressor(random_state=42, verbose=False)
```

- Dalam proses model benchmarking, dilakukan juga validasi data melalui `crossvalidate` dengan metode `Kfold`. crossvalidate dengan metode Kfold dilakukan untuk memvalidasi kinerja model dengan membagi dataset menjadi K subset (dalam data set ini, K = 5 ), yang dikenal sebagai "lipatan" atau "folds," di mana model akan diuji pada satu subset dan dilatih pada K-1 subset lainnya.

- Selain itu, dalam model benchmarking, juga di gunakan `transformed target regression` untuk mengatasi masalah, outlier, ketidaknormalan atau ketidakstabilan variabel target (CLV) yang ada dalam data ini.

```python
from sklearn.model_selection import KFold, cross_validate

models = {
    'Linear Regression': linear_reg,
    'Lasso': lasso,
    'Ridge': ridge,
    'Elastic Net': enet,
    'KNN Regression': knn_reg,
    'Decision Tree Reg': tree_reg,
    'XGBoost Reg':xgb_reg,
    'Random Forest Reg':rf_reg,
    'Gradient Boost Reg':gboost_reg,
    'Catboost Reg': catboost
}

metrics = {'mae': 'neg_mean_absolute_error',
           'mape': 'neg_mean_absolute_percentage_error',
           'rmse': 'neg_root_mean_squared_error'}

name_model = []       
test_mae_arr = []    
test_mae_arr_std = []    
test_mape_arr = []
test_mape_arr_std = []
test_rmse_arr = []
test_rmse_arr_std = []

for name,model in models.items():
    crossval = KFold(n_splits=5)

    estimator = Pipeline([
        ('transform', transformer),
        ('model', model),
    ])

    scores = cross_validate(estimator,X_train,y_train,cv=crossval, scoring=metrics)
    
    name_model.append(name)
    test_mae_arr.append(scores['test_mae'].mean())
    test_mae_arr_std.append(scores['test_mae'].std())
    test_mape_arr.append(scores['test_mape'].mean())
    test_mape_arr_std.append(scores['test_mape'].std())
    test_rmse_arr.append(scores['test_rmse'].mean())
    test_rmse_arr_std.append(scores['test_rmse'].std())

for name,model in models.items():
    crossval = KFold(n_splits=5)
    model_ttr = TransformedTargetRegressor(model, func=np.log, inverse_func=np.exp)

    estimator = Pipeline([
        ('transform', transformer),
        ('model', model_ttr),
    ])

    scores = cross_validate(estimator,X_train,y_train,cv=crossval, scoring=metrics)
    
    name_model.append(f'{name} + TTR')
    test_mae_arr.append(scores['test_mae'].mean())
    test_mae_arr_std.append(scores['test_mae'].std())
    test_mape_arr.append(scores['test_mape'].mean())
    test_mape_arr_std.append(scores['test_mape'].std())
    test_rmse_arr.append(scores['test_rmse'].mean())
    test_rmse_arr_std.append(scores['test_rmse'].std())


model_bench = pd.DataFrame({
    'model': name_model,
    'MAE Mean': test_mae_arr,
    'MAE STD': test_mae_arr_std,
    'MAPE': test_mape_arr,
    'MAPE STD': test_mape_arr_std,
    'RMSE': test_rmse_arr,
    'RMSE STD': test_rmse_arr_std,
})

model_bench
```

```python
model_bench.sort_values('RMSE', ascending=False)
```

- Dari hasil model benchmarking dapat diketahui bahwa model terbaik adalah Gradient Boosting with transform target regression, karena memiliki nilai matriks error (RMSE, MAPE, dan MAE) yang paling kecil. Standar deviasi errornya pun dapat dikatakan stabil.
- matriks RMSE digunakan dalam data ini karena Lebih sensitif terhadap perbedaan dalam kesalahan antara prediksi dan nilai sebenarnya, yang dapat bermanfaat dalam mengidentifikasi pengamatan yang signifikan. RMSLE tidak dapat digunakan dalam data ini karena tidak terlalu cocok untuk data dengan nilai 0 yang banyak (income memiliki nilai 0 yang banyak). sedangkan MAPE digunakna karena lebih mudah di interpretasi dan mudah di jelaskan secara general kepada audience yang tidak terlalu menegerti tentang matriks, sedangkan MAE digunakna karena tidak terlalu sensitif terhadap outliers seperti yang ada di data ini, karena menggunakan nilai absolut kesalahan. 

### Predict to Test Set

```python
gboost_reg = GradientBoostingRegressor(random_state=42)
gb_ttr = TransformedTargetRegressor(gboost_reg, func=np.log, inverse_func=np.exp)


catboost= CatBoostRegressor(random_state=42, verbose=False)
catboost_ttr = TransformedTargetRegressor(catboost, func=np.log, inverse_func=np.exp)

models = [(gb_ttr, "Gradient Boosting + TTR"), (catboost_ttr, "CatBoost +TTR")]

results = []

for model, model_name in models:
    model_pipeline = Pipeline([
        ('transform', transformer),
        ('model', model)
    ])
    
    score_rmse = []
    score_mae = []
    score_mape = []

    model_pipeline.fit(X_train, y_train)
    y_pred = model_pipeline.predict(X_test)
    score_rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    score_mae.append(mean_absolute_error(y_test, y_pred))
    score_mape.append(mean_absolute_percentage_error(y_test, y_pred))

    score_df = pd.DataFrame({'RMSE': score_rmse, 'MAE': score_mae, 'MAPE': score_mape}, index=[model_name])

    results.append(score_df)

combined_results = pd.concat(results)

combined_results
```

Setelah mencoba dua model terbaik di data set test, diketahui bahwa hasil algoritma yang terbaik adalah algoritma Gradient Boosting dengan Transformed Target Regression. Gradient Boosting TTR unggul di dua metriks MAPE dan MAE dibandingkan dengan Catboost.

Gradient Boosting Regressor itu sendiri adalah, salah satu jenis model machine learning yang digunakan untuk regresi, yaitu untuk memprediksi nilai numerik atau kontinu berdasarkan fitur-fitur yang diberikan dalam hal ini adalah mempredisksi nilai customer lifetime value menggunakan fitur-fitur yang ada. Model ini adalah bagian dari keluarga algoritma ensemble learning, yang berarti ia menggabungkan beberapa model lemah (biasanya decision tree yang disebut "weak learners") untuk membuat model yang lebih kuat. Gradient Boosting menggunakan teknik ensemble learning, di mana beberapa pohon keputusan digunakan bersama-sama untuk meningkatkan kinerja prediksi. Pohon-pohon keputusan ini secara bertahap dibangun dan digabungkan untuk mengurangi kesalahan prediksi. Proses training gradient boosting ini berpusat pada konsep "boosting". yang mana pada setiap langkah iterasi, model berusaha untuk mengoreksi kesalahan yang dibuat oleh model sebelumnya. Dan dilakukan dengan menghitung gradien dari kesalahan yang tersisa dan memperbarui model sebelumnya untuk meminimalkan kesalahan tersebut.

Sedangkan Transformed Target Regression, digunakan untuk mengatasi masalah, outlier, ketidaknormalan atau ketidakstabilan variabel target (CLV) yang ada dalam data ini.

Oleh karena itu, untuk selanjutnya di tahap Hyperparameter tuning, model yang digunakan hanya satu yang terbaik yaitu Gradient Boosting Regressor

### `Hyperparameter Tuning`

Hyperparameter tuning adalah proses menyesuaikan atau mengoptimalkan hyperparameter dalam sebuah model atau algoritma machine learning untuk mencapai performa yang lebih baik.

Hyperparameter tuning yang digunakan dalam data ini adalah berjenis `Grid search` yang melibatkan pengujian semua kombinasi hyperparameter yang mungkin dari sekumpulan nilai yang telah ditentukan sebelumnya.

Dalam melakukan proses hyperparameter tuning, hal yang perlu dilakukan pertama-tama adalah menentukan hyperparameter spacenya. 
adapun komponen hyperparameter space yang digunakan dalam data ini adalah:
1. n_estimators= Jumlah pohon keputusan yang akan digunakan dalam proses pembelajaran Gradient Boosting. semakin banyak pohon keputusan yang akan digunakan, yang pada gilirannya dapat meningkatkan kinerja model secara umum
2. learning_rate = Parameter yang digunakan untuk mengontrol sejauh mana setiap pohon keputusan dalam ensemble berkontribusi pada prediksi akhir model. learning rate yang lebih besar akan memberikan kontribusi yang lebih besar dari setiap pohon, yang dapat menghasilkan model yang lebih kompleks.
3. max_depth= maksimal jumlah level cabang dalam pohon keputusan yang menggambarkan tingkat kedalaman pohon keputusan, semakin dalam akan semakin kompleks.
4. min_samples_split=  jumlah minimum sampel yang diperlukan dalam sebuah node untuk memutuskan apakah node tersebut akan dibagi lebih lanjut menjadi cabang-cabang yang lebih kecil atau tidak.
5. min_samples_leaf=  jumlah minimum sampel yang harus ada dalam setiap daun (leaf) dari pohon keputusan.
6. subsample= proporsi sampel pelatihan yang digunakan untuk melatih setiap pohon keputusan dalam ensemble

Dibawah ini adalah kombinasi hyperparameter space yang digunakan.

```python
# untuk estimator

hyperparam_space = {
    'model__regressor__n_estimators': [50, 100, 150],  
    'model__regressor__learning_rate': [0.01, 0.1, 0.5],
    'model__regressor__max_depth': [3, 4, 5],
    'model__regressor__min_samples_split': [2, 3, 4],
    'model__regressor__min_samples_leaf': [1, 2, 3],
    'model__regressor__subsample': [0.8, 0.9, 1.0]
}
```

```python
gboost_reg = GradientBoostingRegressor(random_state=42)
gb_ttr = TransformedTargetRegressor(gboost_reg, func=np.log, inverse_func=np.exp)
model = Pipeline([
    ('transform', transformer),
    ('model', gb_ttr)
])

gridsearch = GridSearchCV(
    estimator=model,          
    param_grid=hyperparam_space,    
    cv=5,                           
    scoring = 'neg_mean_absolute_percentage_error', 
    n_jobs = -1               
)

gridsearch.fit(X_train, y_train)
```

```python
model = gridsearch.best_estimator_
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
score_rmse = []
score_mae = []
score_mape = []
score_rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
score_mae.append(mean_absolute_error(y_test, y_pred))
score_mape.append(mean_absolute_percentage_error(y_test, y_pred))


score_after_tuning = pd.DataFrame({'RMSE': score_rmse, 
                                    'MAE': score_mae, 
                                    'MAPE': score_mape}, 
                                    index=['Gradient Boosting Regressor + TTR after tuning'])
score_after_tuning
```

Untuk mempermudah penjelasan terkait dengan residual plot dan actual vs predict CLV, kita dapat melakukan pengelompokan data melalui binning dengan tujuan untuk mengkelaskan data yang bersifat kontinu menjadi kelompok-kelompok yang lebih kecil atau interval. Tujuan utama dari binning adalah untuk menyederhanakan distribusi data dan membuatnya lebih mudah untuk dianalisis. Dari hasil binning ini akan diketahui kelompok data tertentu dan nilai MAPE nya.

```python
data = pd.DataFrame(y_test).rename(columns={'Customer Lifetime Value':'true'})
data['pred'] = y_pred
data['bins'] = pd.cut(x=data['true'], 
                      bins=[1800, 18000, 36000, 54000, 72000, 90000], 
                      labels=['1800-18000', '18000-36000', '36000-54000', '54000-72000','72000-90000'])
data.tail()
```

```python
from sklearn.metrics import mean_absolute_percentage_error

def mape(data):
    return mean_absolute_percentage_error([data['true']], [data['pred']])

data['mape'] = data[['true','pred']].apply(mape, axis=1)
data.head()
```

```python
data.groupby('bins')[['mape']].mean()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))

sns.countplot(data=data, x='bins', color='lightblue', ax=ax1)
ax1.set_ylabel('count of data')

data_by = data.groupby('bins')[['mape']].mean()
cols = ['palegreen' if x < 0.25 else '#708090' for x in data_by['mape']]
sns.barplot(data=data, x='bins', y='mape', 
            palette=cols, ax=ax2, 
            estimator='mean',
            errorbar=None)
ax2.axhline(y=0.25, color='red', ls='--');
ax2.set_ylabel('% MAPE');
```

Dapat diketahu dari hasil barchart di atas bahwa, semakin besar nilai CLV nya maka semakin sedikit jumlah datanya namun semakin besar nilai MAPE nya. ini berarti error atau spread prediksi dan aktual CLV yang tinggi terjadi karena jumlah data yang tidak merata. ketika nilai CLV berada di rentang 1800-18000, MAPE nya rendah dan juga jumlah datanya banyak.
